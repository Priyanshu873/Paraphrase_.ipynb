```python
model = SentenceTransformer("distiluse-base-multilingual-cased-v1")
```

```python
corpos1 = ["The cat sits outside",
"A man is playing guitar",
"I love pasta",
"The new movie is awesome",
"The cat plays in the garden"]

corpos_emb = model.encode(corpos1, convert_to_tensor=True)
```

```python
cosine_similarity = util.pytorch_cos_sim(corpos_emb, corpos_emb)
avg_sim = cosine_similarity.mean(dim=1)
outlier_index = avg_sim.argmin().item()
outlier_index
print(f"Outlier sentence: {corpos1[outlier_index]}(avg similarity : {avg_sim[outlier_index]})")
```
